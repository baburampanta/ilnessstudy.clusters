```python
#Load Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#load csv file
my_data=pd.read_csv('illnessstudy.csv')
my_data.head()
```

```python
#Create x (we ignore the y variable)
x = my_data.drop('diagnosis', axis=1).to_numpy()

#Scale the Data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data_transformed=sc.fit_transform(x)
```

```python
#Determine optimum number of clusters
wcss = []
for i in range(1, 10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10,random_state = 100)
    kmeans.fit(data_transformed)
    wcss.append(kmeans.inertia_)
    
#Plot Elbow Method
plt.plot(range(1, 10), wcss,marker='o')
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create Silhouette Coefficients
scores = [0] 
for n_cluster in range(2, 10):
    kmeans = KMeans(n_clusters=n_cluster).fit(data_transformed)  # Use the scaled data
    labels = kmeans.labels_
    sil_coeff = silhouette_score(data_transformed, labels, metric='euclidean')  # Calculate silhouette coefficient
    scores.append(sil_coeff) 
    print(f'For n_clusters = {n_cluster}, The Silhouette Coefficient is {sil_coeff}')

# Plot Silhouette Method
plt.plot(range(2, 10), scores[1:], marker='o')  # Start plotting from n_clusters=2
plt.title('The Silhouette Method')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Coefficient')
plt.show()
```

```python
#Create KMeans with 2 Clusters
my_data1= my_data.drop('diagnosis', axis=1)
# Apply KMeans clustering
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=300, n_init=10, random_state=100)
y_kmeans = kmeans.fit_predict(my_data1)

# Add the predicted cluster labels to the dataset
my_data1['New Cluster'] = kmeans.labels_

# Map the numeric labels to more readable cluster names
my_data1['New Cluster'] = my_data1['New Cluster'].map({0: 'Cluster 1', 1: 'Cluster 2'})

# Display the first few rows of the dataset with the new cluster labels
print(my_data1.head())
```

```python
# Plot of 2 Clusters
plt.scatter(data_transformed[y_kmeans == 0, 0], data_transformed[y_kmeans == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(data_transformed[y_kmeans == 1, 0], data_transformed[y_kmeans == 1, 1], s = 100, c = 'pink', label = 'Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'blue', label = 'Centroids')

# Add title and labels
plt.title('KMeans Clustering with 2 Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Display the legend outside the plot
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)

# Show the plot
plt.show()
```

```python
# Apply KMeans clustering with 3 clusters
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 100)
y_kmeans = kmeans.fit_predict(data_transformed)

# Plot of 3 Clusters
plt.figure(figsize=(10, 6))  # Optional: Adjust the figure size
plt.scatter(data_transformed[y_kmeans == 0, 0], data_transformed[y_kmeans == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(data_transformed[y_kmeans == 1, 0], data_transformed[y_kmeans == 1, 1], s = 100, c = 'red', label = 'Cluster 2')
plt.scatter(data_transformed[y_kmeans == 2, 0], data_transformed[y_kmeans == 2, 1], s = 100, c = 'blue', label = 'Cluster 3')

# Plot centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids', marker='X')

# Set plot labels and title
plt.title('KMeans Clustering (3 Clusters)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Show legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='best')
plt.tight_layout()  # Optional: Adjust layout to avoid overlap
plt.show()
```
